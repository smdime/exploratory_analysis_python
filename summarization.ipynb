```python
import pandas as pd
```

```python
df = pd.read_csv("sample_dataset.csv")
```

# Summarization is a method that gives a quick metrics so we can calculate the dataset feature by feature.
The mean value for each feature, the standard deviation, the percentiles, the maximum and minimum values, etc.
Summarization gives a brief summary of each feature giving us the basic statistics properties of this feature.
We can summariz our dataset using the describe method.

```python
df.describe() #view the statistics of the dataset
```

```python
df.describe(percentiles=[0.25, 0.5, 0.75, 0.05, 0.95]) #calculate the different summry
#percentile give an overview of the shape of the distribution each feature
```

```python
df.describe(include=['object','category']) #categorical values, no category
```

```python
df['area error'].value_counts() #to look at the counts
```
