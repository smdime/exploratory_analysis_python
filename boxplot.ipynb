```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("sample_dataset.csv")
```

```python
df.columns
```

```python
df[['mean radius']].boxplot()
```

- boxplot is a method in pandas dataframe therefore needs double square brackets
- 1st quartile (lower line)
- 3rd quartile (upper line)
- 2nd quartile (median)

IQR = interquartile range
 = difference of the 3rd and 1st quartiles

upper outer line = 3rd quartile + IQR * 1.5
lower outer line = 1st quartile - IQR * 1.5

outliers = points outside the upper outer line
